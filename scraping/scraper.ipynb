```python
import requests
import ssl
import numpy as np
import urllib
import os
```

```python
ASOS_URL = "https://api.asos.com/product/search/v1/?q={}&store=1&lang=en-GB&sizeschema=EU&currency=EUR&sort=freshness&channel=mobile-app&offset=0&limit=100"
hdr = {'User-Agent': 'Mozilla/5.0'}
```

```python
res = requests.get(ASOS_URL.format("jeans"), timeout=2, headers=hdr)
```

```python
def fetch_(_id):
    url = ASOS_URL.format(_id)
    res = requests.get(url, headers=hdr, timeout=10)
    return res.json()['products']
```

```python
## get item categories to id img dict
item_imgs = {}
raw_data = {}
categories = ["jeans", "shorts", "skirts", "jumper", "loungewear", "hoodies", "dresses"]
for item in categories:
    data = fetch_(item)
    raw_data[item] = data
    item_imgs[item] = {i['id']: i['images'] for i in data}
```

```python
## save imgs from url and return fp 
def url_to_img_file(img_url, _dir, _id):
    img_root = os.path.join(os.getcwd(), "asos_imgs")
    _dir = os.path.join(img_root, _dir)
    fp = os.path.join(_dir, "{}.png".format(_id))
    if img_url:
        urllib.request.urlretrieve(img_url, fp)
        return fp
    else:
        return None
```

```python
def get_imgs(data, _type):
    for i in data.keys():
        for img in data[i]:
            fp = url_to_img_file(img['url'], _type, i)
            img['url'] = fp
    return data
```

```python
for key, products in item_imgs.items():
    item_imgs[key] = get_imgs(products, key)
```

```python
import pickle as pkl
```

```python
with open("img_raw_data.pkl", "wb") as pfile:
    pkl.dump(raw_data, pfile)
```

```python
with open("item_imgs.pkl", "wb") as pfile:
    pkl.dump(item_imgs, pfile)
```
